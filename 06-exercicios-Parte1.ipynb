```python
vogais = "aeiouAEIOU"
novo_texto = ""
for letra in String:
  if letra in vogais:
    novo_texto += "*"
  else:
    novo_texto += letra
    
print(novo_texto)
```

```python
palavras = String.split()
palavras_maisculas = [palavra for palavra in palavras if palavra[0].isupper()]
texto_maiusculo = "".join(palavras_maisculas)
print(texto_maiusculo)
```

### 5. Crie um programa em Python que receba uma string como entrada e verifique se ela é um palíndromo (ou seja, se pode ser lida da mesma forma da esquerda para a direita e da direita para a esquerda).

```python
texto = input("Digite uma string: ")
texto_invertida = texto[::-1]

texto_novo = "".join(texto)
texto_invertida_novo = "".join(texto_invertida)

if texto_novo == texto_invertida_novo:
  print('É um Palíndromo')      
else:
  print('Não é um Palíndromo') 
```

### 6. Crie um programa em Python que receba uma string como entrada e verifique se ela é um anagrama de outra string. (ou seja, se elas possuem as mesmas letras em quantidades iguais)

```python
texto1 = input("Digite a primeira string: ")
texto2 = input("Digite a segunda string: ")

primeira_string = texto1.upper()
segunda_string = texto2.upper()

primeira_string = primeira_string.replace(" ","")
segunda_string = segunda_string.replace(" ","")

if sorted(primeira_string) == sorted(segunda_string):
  print('São Anagramas')
  
else:
  print('Não São Anagramas')


```
