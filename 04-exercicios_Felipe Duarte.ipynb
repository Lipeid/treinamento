### 1. Some todos os números de 1 até 100

```python
numeros = range(1,101)

for numero in numeros:
    print(numero, end=' ')
  
```

```python
soma = 0
for numero in numeros:
    if numero <= 101:
        soma = soma + numero
        
print(soma)
 
```

### 2. Some todos os números pares de 1 até 100

```python
numeros = range(1,101)

for numero in numeros:
    print(numero, end=' ')
```

```python
for numero in numeros:
    if numero % 2 == 0:
        print(numero, end=' ')
```

```python
soma = 0
for numero in numeros:
    if numero % 2 == 0:
       soma = soma + numero
      
print(soma)
```

### 3. Some todos os números ímpares de 1 até 100

```python
numeros = range(1,101)

for numero in numeros:
    print(numero, end=' ')
```

```python
for numero in numeros:
    if numero % 2 != 0:
        print(numero, end=' ')
```

```python
soma = 0

for numero in numeros:
    if numero % 2 != 0:
        soma = soma + numero
        
print(soma)
```
